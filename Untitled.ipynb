```python
# Initial imports
import pandas as pd
from pathlib import Path
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Data loading
file_path = Path("./Resources/myopia.csv")
myopia = pd.read_csv(file_path)
myopia.head(10)
```

```python
myopia_df= myopia.drop("MYOPIC", axis =1)
myopia_df.head()
```

```python
# Standarize data with StandarScaler
myopia_scaled = StandardScaler().fit_transform(myopia_df)
print(myopia_scaled[0:5])
```

```python
# Part 2
# Perfrom dimensionality reduciton with PCA

# Initialize PCA model
pca = PCA(n_components=2)


myopia_pca = pca.fit_transform(myopia_scaled)

myopia_df_pca = pd.DataFrame(
    data=myopia_pca, columns=["principal component 1", "principal component 2"]
)
myopia_df_pca.head()
```

```python
tsne = TSNE(learning_rate=100)
transformed_features = tsne.fit_transform(myopia_pca)
```

```python
myopia_df["x"] = transformed_features[:,0]
myopia_df["y"] = transformed_features[:,1]
myopia_df["class"] = myopia["MYOPIC"]
```

```python
plt.scatter(myopia_df["x"], myopia_df["y"])
plt.show()
```

```python
plt.scatter(myopia_df["x"], myopia_df["y"], c=myopia_df["class"])
plt.show()
```
